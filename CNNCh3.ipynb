```python
import tensorflow as tf

data = tf.keras.datasets.fashion_mnist
(training_images, training_labels), (test_images, test_labels) = data.load_data()

training_images = training_images.reshape(60000, 28, 28, 1)
training_images = training_images / 255.0
test_images = test_images.reshape(10000, 28, 28, 1)
test_images = test_images / 255.0

class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('accuracy')>0.95):
            print("\nReached 95% accuracy so cancelling training!")
            self.model.stop_training = True

callbacks = myCallback()

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', input_shape=(28, 28, 1)), # convolutional layer with 64 3X3 filters
    tf.keras.layers.MaxPooling2D(2, 2), # Pooling layer to reduce pixels
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'), # another convolutional layer
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'), # another convolutional layer
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(), # Flatten layer to turn into 1D for the dense layer
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(64, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])

model.compile(optimizer='adam' ,loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(training_images, training_labels, epochs=30, callbacks = [callbacks])
model.evaluate(test_images, test_labels)
```

```python
import numpy as np
classifications = model.predict(test_images)
print(np.argmax(classifications[0]))
print(test_labels[0])
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sanikamal/horses-or-humans-dataset")

print("Path to dataset files:", path)
```

```python
import os
# define training and validation data directories
data_dir = '/kaggle/input/horses-or-humans-dataset/horse-or-human/'
train_dir = os.path.join(data_dir, 'train/')
validation_dir = os.path.join(data_dir, 'validation/')

# classes are folders in each directory with these names
classes = ['horse','human']
```

```python
## Defining the model and impolelmenting
import tensorflow as tf
from tensorflow.keras.optimizers import RMSprop

model2 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', input_shape=(300, 300, 3)), # convolutional layer with 64 3X3 filters
    tf.keras.layers.MaxPooling2D(2, 2), # Pooling layer to reduce pixels
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'), # another convolutional layer
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(256, (3, 3), activation='relu'), # another convolutional layer
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(), # Flatten layer to turn into 1D for the dense layer
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(64, activation=tf.nn.relu),
    tf.keras.layers.Dense(1, activation='sigmoid')]) # one neuron with sigmoid activation for binary classification
    # value close to 0 for one class and value close to 1 for the other

model2.compile(optimizer = RMSprop(learning_rate = 0.001) ,loss='binary_crossentropy', metrics=['accuracy'])
# using binary crosse entropy loss function because we are dealing with binary classification
# this RMS optimizer works better for binary classification

model2.summary()
```

```python
## Keras ImageDataGenerator, data augmentation and validation set example

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1/255,
                                   rotation_range = 40,
                                   width_shift_range = 0.2,
                                   height_shift_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True,
                                   fill_mode = 'nearest') # fill any missing pixels(due to zoom or rotation) with the values of their nearest pixels

validation_datagen = ImageDataGenerator(rescale = 1/255,
                                   rotation_range = 40,
                                   width_shift_range = 0.2,
                                   height_shift_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True,
                                   fill_mode = 'nearest')

train_generator = train_datagen.flow_from_directory(train_dir, target_size = (300,300), class_mode = 'binary')
validation_generator = train_datagen.flow_from_directory(validation_dir, target_size = (300,300), class_mode = 'binary')

```

```python
history = model2.fit(train_generator,
                    epochs=10,
                    validation_data=validation_generator)
```

```python
# Wow, the previous model is HORRIBLE
## Transfer Learning and dropout Example
import urllib.request

from tensorflow.keras.applications.inception_v3 import InceptionV3

weights_url = "https://storage.googleapis.com/tensorflow/keras-applications/inception_v3/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5"

weights_file = "inception_v3.h5"

# Downloading the weights of the inception model from google
urllib.request.urlretrieve(weights_url, weights_file)

# Defining the model architecture with the downloaded weights
pre_trained_model = InceptionV3(input_shape=(150, 150, 3),include_top=False,weights=None)
pre_trained_model.load_weights(weights_file)

pre_trained_model.summary()
```

```python
# Make all the features of the pre trained model untrainable
# Because we are using the already trained weights to train another model
for layer in pre_trained_model.layers:
  layer.trainable = False

# placcing the ast layer into a variable
# This is the layer we want to be our last layer before we add our dense layer
last_layer = pre_trained_model.get_layer('mixed7')
print('last layer output shape: ', last_layer.output.shape)

last_output = last_layer.output
```

```python
from tensorflow.keras import layers, Model

# Flatten the output layer to 1 dimension
x = layers.GlobalAveragePooling2D()(last_output)
# Add a fully connected layer with 1,024 hidden units and ReLU activation
x = layers.Dense(1024, activation='relu')(x)
# Add a dropout layer to avoid overfitting
# This is done by randommly switching off neurons during training which avoids them from contributing to the network
x = layers.Dropout(0.2)(x)
# Add a final sigmoid layer for classification
x = layers.Dense(1, activation='sigmoid')(x)
```

```python
train_generator = train_datagen.flow_from_directory(train_dir, target_size = (150,150), class_mode = 'binary')
validation_generator = train_datagen.flow_from_directory(validation_dir, target_size = (150,150), class_mode = 'binary')
```

```python
## Defining the model using the pretrained model and the other layers we added
model = Model(pre_trained_model.input, x)

model.compile(optimizer=RMSprop(learning_rate = 0.0001), loss='binary_crossentropy',metrics=['acc'])

# Fit the model
history = model.fit(train_generator, epochs = 10, validation_data = validation_generator)
```
