```python
import numpy as np
import pandas as pd
import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
```

```python
window_size = 10
sentences = []

data = open("Jcole.txt", encoding="utf-8").read()
corpus = data.lower()
words = corpus.split()

range_size = len(words) - window_size

for i in range(range_size):
    thissentence = " ".join(words[i:i + window_size])
    sentences.append(thissentence)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
total_words = len(tokenizer.word_index) + 1
```

```python
input_sequences = tokenizer.texts_to_sequences(sentences)
```

```python
X = []
y = []

for seq in input_sequences:
    X.append(seq[:-1])
    y.append(seq[-1])

maxlen = max(len(seq) for seq in X)
X = pad_sequences(X, maxlen=maxlen, padding='pre')  # 'pre' means pad at the start

X = np.array(X)
y = np.array(y)
```

```python
from keras.utils import to_categorical

y = to_categorical(y, num_classes=total_words)
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Bidirectional

model = Sequential()
model.add(Embedding(total_words, 8))
model.add(Bidirectional(LSTM(maxlen-1, return_sequences='True')))
model.add(Bidirectional(LSTM(maxlen-1)))
model.add(Dense(total_words, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X, y, epochs=1000, verbose=1)
```
